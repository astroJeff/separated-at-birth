### Load the data

```python
import os
import emcee
import triangle
from scipy.stats import truncnorm, norm
from scipy.interpolate import interp1d
import scipy.optimize as opt
import matplotlib.pyplot as plt
import numpy as np
import pylab as P
%matplotlib inline


# Import MESA lifetimes
datafile = os.path.abspath("../data/test_lifetime.dat")
#datafile = os.path.abspath("../data/MESA_lifetime.dat")
names = ["name","lifetime"]
MESA_tcool = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)

# Move to log10 of stellar ages
MESA_tcool["lifetime"] = np.log10(MESA_tcool["lifetime"])

    
```

### Create function to interpolate between MESA lifetimes

```python
# Function to interpolate MESA lifetimes
def lifetime(MESA_tcool, M_zams):
    
    lifetime = interp1d(MESA_tcool["name"],pow(10.,MESA_tcool["lifetime"]), kind='linear', bounds_error=False, fill_value=1.0e7)
    
    return lifetime(M_zams)


# Function to interpolate M_zams from MESA lifetime (inverse of previous function)
def lifetime_inv(MESA_tcool, tcool):

    # For whatever reason, interp1d requires an increasing function,
    # therefore we create a "view" into the function using the [::-1]
    lifetime_inv = interp1d(pow(10.,MESA_tcool["lifetime"][::-1]),MESA_tcool["name"][::-1], kind='linear', bounds_error=False, fill_value = 0.45)
    
    return lifetime_inv(tcool)


# Function to return M_zams from M_wd using model
def get_mzams(p, m_wd):
    bt1, th1, th2, th3 = p
    
    # Get slopes from theta's
    m1 = np.tan(th1)
    m2 = np.tan(th2)
    m3 = np.tan(th3)
    
    # Determine b2, b3 as a function of other parameters
    b1 = bt1 / np.cos(th1)
    b2 = (m1-m2) * 2.0 + b1
    b3 = (m2-m3) * 4.0 + b2
    
    # Define array
    M_zams = np.array([])

    # Define piecewise linear model
    conds = [m_wd<b1, (m_wd>b1) & (m_wd<2.0*m1+b1), (m_wd>2.0*m1+b1) & (m_wd<4.0*m2+b2), m_wd>4.0*m2+b2]
    funcs = [lambda m_wd: 0., lambda m_wd: 1./m1*(m_wd-b1), lambda m_wd: 1./m2*(m_wd-b2), lambda m_wd: 1./m3*(m_wd-b3)]

    M_zams = np.piecewise(m_wd, conds, funcs)
    
    return M_zams


# Function to return M_zams from M_wd using model
def get_m_wd(p, m_zams):
    bt1, th1, th2, th3 = p
#    bt1, th1 = p
    
    # Get slopes from theta's
    m1 = np.tan(th1)
    m2 = np.tan(th2)
    m3 = np.tan(th3)
    
    # Determine b2, b3 as a function of other parameters
    b1 = bt1 / np.cos(th1)
    b2 = (m1-m2) * 2.0 + b1
    b3 = (m2-m3) * 4.0 + b2
    
    # Define array
    M_wd = np.array([])

    # Define piecewise linear model
    conds = [m_zams<0.0, (m_zams>0.0) & (m_zams<2.0), (m_zams>2.0) & (m_zams<4.0), m_zams>4.0]
    funcs = [lambda m_zams: 0., lambda m_zams: m1*m_zams+b1, lambda m_zams: m2*m_zams+b2, lambda m_zams: m3*m_zams+b3]
    
    M_wd = np.piecewise(m_zams, conds, funcs)
    
    return M_wd



```

### Generate Fake Data

```python
def create_mock_data(p_true, N_sys=20):

    # Get random WD masses
    lower = 0.55
    upper = 1.2
    mu = 0.65
    sigma = 0.15
    M1_wd = truncnorm.rvs((lower-mu)/sigma,(upper-mu)/sigma,loc=mu,scale=sigma,size=N_sys)

    lower = 0.45
    upper = 1.0
    mu = 1.0
    sigma = 0.10
    mass_ratio = truncnorm.rvs((lower-mu)/sigma,(upper-mu)/sigma,loc=mu,scale=sigma,size=N_sys)
    M2_wd = M1_wd * mass_ratio

    # Determine their ZAMS masses
    M1_zams = get_mzams(p_true, M1_wd)
    M2_zams = get_mzams(p_true, M2_wd)

    # Get the Main Sequence lifetimes from these ZAMS masses
    M1_lifetime = lifetime(MESA_tcool, M1_zams)
    M2_lifetime = lifetime(MESA_tcool, M2_zams)
    delta_MS_lifetime = M2_lifetime-M1_lifetime

    # Get random system ages: log t: Uniform (7.0,9.0)
    age_system = np.power(10.0, np.random.uniform(7.0,9.0,N_sys))
    t1_cool = age_system + delta_MS_lifetime
    t2_cool = age_system 

    # Add fake uncertainties: 10% for cooling age, 0.3 Msun in mass
    t1_cool = np.random.normal(t1_cool, 0.1*t1_cool)
    t2_cool = np.random.normal(t2_cool, 0.1*t2_cool)
    M1_wd = np.random.normal(M1_wd,0.03)
    M2_wd = np.random.normal(M2_wd,0.03)

    M1_wd_err = 0.03*np.ones(N_sys)
    M2_wd_err = 0.03*np.ones(N_sys)
    t1_cool_err = 0.1 * t1_cool
    t2_cool_err = 0.1 * t2_cool

    # Move data into structure "DWD"
    dtype=[('m1_wd','float64'),('m1_wd_err','float64'),('m2_wd','float64'),('m2_wd_err','float64'),
                      ('age1','float64'),('age1_err','float64'),('age2','float64'),('age2_err','float64')]

    DWD = np.zeros(N_sys, dtype=dtype)
    DWD["m1_wd"] = np.copy(M1_wd)
    DWD["m1_wd_err"] = np.copy(M1_wd_err)
    DWD["m2_wd"] = np.copy(M2_wd)
    DWD["m2_wd_err"] = np.copy(M2_wd_err)
    DWD["age1"] = np.copy(t1_cool)
    DWD["age1_err"] = np.copy(t1_cool_err)
    DWD["age2"] = np.copy(t2_cool)
    DWD["age2_err"] = np.copy(t2_cool_err)

    return DWD
```

### Functions for likelihood, prior, posterior, etc. for 3-piecewise linear model

```python
# Function for optimization
def temp_func(x, delta_time, delta_cool_err):
    arg = np.exp(-(delta_time - x)*(delta_time - x)/(2. * delta_cool_err*delta_cool_err))
    arg[np.isinf(arg)] = 0.0
#    print x, -np.sum(arg)
    return -np.sum(arg)

def temp_func_deriv(x, delta_time, delta_cool_err):
#    a = (temp_func(x+100.0, delta_time, delta_cool_err) - temp_func(x, delta_time, delta_cool_err))/100.0    
    arg = np.exp(-(delta_time - x)*(delta_time - x)/(2. * delta_cool_err*delta_cool_err))
    arg[np.isinf(arg)] = 0.0
#    print a, -np.sum((delta_time - x)/(delta_cool_err*delta_cool_err) * arg)
    return -np.sum((delta_time - x)/(delta_cool_err*delta_cool_err) * arg)
#    return a


# Calculate likelihood function

ln_P_sys = np.array([])
def ln_likelihood(p, MESA_tcool, DWD):
    

    # Define lifetime interpolation functions
    func_lifetime = interp1d(MESA_tcool["name"],pow(10.,MESA_tcool["lifetime"]), kind='linear', bounds_error=False, fill_value=1.0e7)    
    func_lifetime_inv = interp1d(pow(10.,MESA_tcool["lifetime"][::-1]),MESA_tcool["name"][::-1], kind='linear', bounds_error=False, fill_value = 0.45)
    
    
    # Number of samples for our Monte Carlo integration
    N_samp = 10192
    
    # Limit the WD mass to be confined within this range
    wd_lower = 0.5    # Lowest allowed WD mass
    wd_upper = 1.4    # Highest allowed WD mass
    # First determine upper and lower
    zams_limits = np.array([min(MESA_tcool["name"]), max(MESA_tcool["name"])])
    wd_limits = get_m_wd(p, zams_limits)
    t_limits = func_lifetime(zams_limits)
    lower = wd_limits[0]
    upper = wd_limits[1]


    # Create probability array
    P_sys = np.array([])
    dist_wd1 = np.array([])
    dist_wd2 = np.array([])

       
    # Calculate each system individually
    for sys in DWD:

        
        # Define array of cooling age differences
        delta_cool = sys["age2"] - sys["age1"]
        delta_cool_err = np.sqrt(sys["age1_err"]*sys["age1_err"] + sys["age2_err"]*sys["age2_err"])

        # Define [truncated] normal distributions for WD masses
        dist_wd1 = truncnorm.rvs((lower - sys["m1_wd"]) / sys["m1_wd_err"], (upper - sys["m1_wd"]) / sys["m1_wd_err"], loc=sys["m1_wd"], scale=sys["m1_wd_err"], size=N_samp)
        dist_wd2 = truncnorm.rvs((lower - sys["m2_wd"]) / sys["m2_wd_err"], (upper - sys["m2_wd"]) / sys["m2_wd_err"], loc=sys["m2_wd"], scale=sys["m2_wd_err"], size=N_samp)    
    
        # Calculate zams masses
        dist_zams1 = get_mzams(p, dist_wd1)
        dist_zams2 = get_mzams(p, dist_wd2)
        
        # Calculate MS lifetimes
        dist_time1 = func_lifetime(dist_zams1)
        dist_time2 = func_lifetime(dist_zams2)

        
        # To adjust based on jacobian
        dm_wd_dm_zams1 = 1.0e-4 / (get_mzams(p, dist_wd1 + 1.0e-4) - get_mzams(p, dist_wd1))
        dm_zams_dt_ms1 = (func_lifetime_inv(dist_time1 + 1.0e4) - func_lifetime_inv(dist_time1)) / 1.0e4

        dm_wd_dm_zams2 = 1.0e-4 / (get_mzams(p, dist_wd2 + 1.0e-4) - get_mzams(p, dist_wd2))
        dm_zams_dt_ms2 = (func_lifetime_inv(dist_time2 + 1.0e4) - func_lifetime_inv(dist_time2)) / 1.0e4
        coeff_norm = (N_samp + 0.0) / np.sum(dm_zams_dt_ms1*dm_wd_dm_zams1 * dm_zams_dt_ms2*dm_wd_dm_zams2)
        

        # Now compare difference in lifetimes to difference in cooling ages
        delta_time = dist_time1 - dist_time2

        avg_delta_time = np.sum(delta_time) / (N_samp + 0.0)

        # To find the maximum likelihood delta_cool
#        meth = 'BFGS'
#        meth = 'CG'
        meth = 'Powell'
#        meth = 'Nelder-Mead'
        result = opt.minimize(temp_func, delta_cool, args=(delta_time, delta_cool_err), method=meth, tol=1.0e-5)
#        result = opt.minimize(temp_func, delta_cool, args=(delta_time, delta_cool_err), method=meth, tol=1.0e-4, jac=temp_func_deriv)    
#        xmin = opt.fmin_cg(temp_func, 0.0, args=(delta_time, delta_cool_err), maxiter = 10, full_output=0, disp=1, gtol=1.0e-8)


        xmin = result.x
        
        # P( delta t_cool | M1_zams_t, M2_zams_t)
        coeff = 1./(delta_cool_err * np.sqrt(2. * np.pi))
        argument = np.exp(-(delta_time - delta_cool)*(delta_time - delta_cool)/(2. * delta_cool_err*delta_cool_err))
        argument2 = np.exp(-(delta_time - xmin)*(delta_time - xmin)/(2. * delta_cool_err*delta_cool_err))
          
        # Priors on the hyperparameters
        P_sys_prior = coeff_norm * (dm_zams_dt_ms1 * dm_zams_dt_ms2 * dm_wd_dm_zams1 * dm_wd_dm_zams2)
#        P_sys_prior = coeff_norm * (dm_zams_dt_ms1 * dm_zams_dt_ms2)

    
        # This is the value of the integrand
        P_sys_temp = P_sys_prior * coeff * argument
        P_sys_temp2 = P_sys_prior * coeff * argument2
        P_sys_temp[np.isinf(P_sys_temp)] = 0.0
        P_sys_temp2[np.isinf(P_sys_temp2)] = 0.0

         
        P_sys = np.append(P_sys, np.sum(P_sys_temp) / np.sum(P_sys_temp2))
#        P_sys = np.append(P_sys, np.sum(P_sys_temp) / (N_samp + 0.0))
                        
    
        
    ln_P_sys = np.log(P_sys)            
    ln_P_sys[np.isnan(ln_P_sys)] = -np.inf
    
    
    return ln_P_sys 



# The prior functions
def ln_prior(p):
    bt1, th1, th2, th3 = p

    b1 = bt1 / np.cos(th1)

    # Get slopes from theta's
    m1 = np.tan(th1)
    m2 = np.tan(th2)
    m3 = np.tan(th3)
    
    # Determine b2, b3 as a function of other parameters
    b2 = (m1-m2) * 2.0 + b1
    b3 = (m2-m3) * 4.0 + b2
    
    lp = 0.
    
    # Limit the x-intercept: b = [0,0.5]
    if b1 < 0. or b1 > 0.5:
        return -np.inf
    
    # The slopes must all be positive
    # This corresponds to theta = [0,pi/2]
    if th1 < 0. or th1 > np.pi/2.:
        return -np.inf
    if th2 < 0. or th2 > np.pi/2.:
        return -np.inf
    if th3 < 0. or th3 > np.pi/2.:
        return -np.inf
    
    
    # Require that M_zams=7 becomes a WD (M_wd < 1.4)
    if 7.0*m3+b3 > 1.4 or 7.0*m3+b3 < 1.0:
        return -np.inf
    
    
    return lp


# The posterior function calls the likelihood and prior functions
def ln_posterior(p, MESA_tcool, DWD):
    
    # Check prior on model first
    lp = ln_prior(p)
    
    if np.any(np.isnan(lp)):
        return -np.inf

    # Then check likelihood of parameters
    ll = ln_likelihood(p, MESA_tcool, DWD)
    
    if np.any(np.isnan(ll) | np.isinf(ll)):
        return -np.inf

    
    if np.all(ll == 0.):
        return -np.inf
    
    return lp + ll.sum()

```

### Wrapper for emcee

```python
p_i = np.copy(p_true)

# This function runs emcee
def run_inference(MESA_tcool, DWD, nwalkers=32, nburn=200, nsteps=1000):
    # number of parameters
    ndim = 4

    # initial positions for walkers
    p0 = np.zeros((nwalkers,ndim))
    p0[:,0] = np.random.normal(p_i[0], 0.01, size=nwalkers) # b1
    p0[:,1] = np.random.normal(p_i[1], 0.01, size=nwalkers) # theta_1
    p0[:,2] = np.random.normal(p_i[2], 0.03, size=nwalkers) # theta_2
    p0[:,3] = np.random.normal(p_i[3], 0.03, size=nwalkers) # theta_3
    
    args = (MESA_tcool, DWD)
    sampler = emcee.EnsembleSampler(nwalkers=nwalkers, dim=ndim, lnpostfn=ln_posterior, args=args)

    # burn-in
    sampler.run_mcmc(p0, N=nburn)
    pos,prob,state = sampler.run_mcmc(p0, N=nburn)

    # throw out burn-in samples
    sampler.reset() 
    pos,prob,state = sampler.run_mcmc(pos, N=nsteps)

    
    return sampler
```

### Run emcee

```python
p_true_1 = np.array([0.1, 0.1, 0.38, 0.05])
N_sys_1 = 20
DWD_1 = create_mock_data(p_true_1, N_sys_1)
sampler = run_inference(MESA_tcool, DWD_1, nburn=200, nsteps=1000)
print "Mock data set 1 finished"

p_true_2 = np.array([0.1, 0.1, 0.38, 0.05])
N_sys_2 = 50
DWD_2 = create_mock_data(p_true_2, N_sys_2)
sampler = run_inference(MESA_tcool, DWD_2, nburn=200, nsteps=1000)
print "Mock data set 2 finished"

p_true_3 = np.array([0.096, 0.291, 0.10, 0.08])
N_sys_3 = 20
DWD_3 = create_mock_data(p_true_3, N_sys_3)
sampler = run_inference(MESA_tcool, DWD_3, nburn=200, nsteps=1000)
print "Mock data set 3 finished"

p_true_4 = np.array([0.096, 0.291, 0.10, 0.08])
N_sys_4 = 50
DWD_4 = create_mock_data(p_true_4, N_sys_4)
sampler = run_inference(MESA_tcool, DWD_4, nburn=200, nsteps=1000)
print "Mock data set 4 finished"

```
