### Create function to interpolate between MESA lifetimes

```python
# Function to interpolate MESA lifetimes
def lifetime(MESA_tcool, M_zams):
    
    lifetime = interp1d(MESA_tcool["name"],pow(10.,MESA_tcool["lifetime"]), kind='linear', bounds_error=False, fill_value=1.0e7)
    
    return lifetime(M_zams)


# Function to interpolate M_zams from MESA lifetime (inverse of previous function)
def lifetime_inv(MESA_tcool, tcool):

    # For whatever reason, interp1d requires an increasing function,
    # therefore we create a "view" into the function using the [::-1]
    lifetime_inv = interp1d(pow(10.,MESA_tcool["lifetime"][::-1]),MESA_tcool["name"][::-1], kind='linear', bounds_error=False, fill_value = 0.45)
    
    return lifetime_inv(tcool)


# Function to return M_zams from M_wd using model
def get_mzams(p, m_wd):
    bt1, th1, th2, th3 = p
    
    # Get slopes from theta's
    m1 = np.tan(th1)
    m2 = np.tan(th2)
    m3 = np.tan(th3)
    
    # Determine b2, b3 as a function of other parameters
    b1 = bt1 / np.cos(th1)
    b2 = (m1-m2) * 2.0 + b1
    b3 = (m2-m3) * 4.0 + b2
    
    # Define array
    M_zams = np.array([])

    # Define piecewise linear model
    conds = [m_wd<=b1, (m_wd>b1) & (m_wd<=2.0*m1+b1), (m_wd>2.0*m1+b1) & (m_wd<=4.0*m2+b2), m_wd>4.0*m2+b2]
    funcs = [lambda m_wd: 0., lambda m_wd: 1./m1*(m_wd-b1), lambda m_wd: 1./m2*(m_wd-b2), lambda m_wd: 1./m3*(m_wd-b3)]

    M_zams = np.piecewise(m_wd, conds, funcs)
    
    return M_zams


# Function to return M_zams from M_wd using model
def get_m_wd(p, m_zams):
    bt1, th1, th2, th3 = p
#    bt1, th1 = p
    
    # Get slopes from theta's
    m1 = np.tan(th1)
    m2 = np.tan(th2)
    m3 = np.tan(th3)
    
    # Determine b2, b3 as a function of other parameters
    b1 = bt1 / np.cos(th1)
    b2 = (m1-m2) * 2.0 + b1
    b3 = (m2-m3) * 4.0 + b2
    
    # Define array
    M_wd = np.array([])

    # Define piecewise linear model
    conds = [m_zams<0.0, (m_zams>=0.0) & (m_zams<2.0), (m_zams>=2.0) & (m_zams<4.0), m_zams>=4.0]
    funcs = [lambda m_zams: 0., lambda m_zams: m1*m_zams+b1, lambda m_zams: m2*m_zams+b2, lambda m_zams: m3*m_zams+b3]
    
    M_wd = np.piecewise(m_zams, conds, funcs)
    
    return M_wd



```

### Functions for likelihood, prior, posterior, etc. for 3-piecewise linear model

```python
# Define lifetime interpolation functions
func_lifetime = interp1d(MESA_tcool["name"],pow(10.,MESA_tcool["lifetime"]), kind='linear', bounds_error=False, fill_value=1.0e7)    
func_lifetime_inv = interp1d(pow(10.,MESA_tcool["lifetime"][::-1]),MESA_tcool["name"][::-1], kind='linear', bounds_error=False, fill_value = 0.45)


# Function for optimization
def temp_func(x, delta_time, delta_cool_err, jacobian):
    arg = jacobian * np.exp(-(delta_time - x)*(delta_time - x)/(2. * delta_cool_err*delta_cool_err))
    arg[np.isinf(arg)] = 0.0
#    print x, -np.sum(arg)
    return -np.sum(arg)

def temp_func_deriv(x, delta_time, delta_cool_err):
#    a = (temp_func(x+100.0, delta_time, delta_cool_err) - temp_func(x, delta_time, delta_cool_err))/100.0    
    arg = np.exp(-(delta_time - x)*(delta_time - x)/(2. * delta_cool_err*delta_cool_err))
    arg[np.isinf(arg)] = 0.0
#    print a, -np.sum((delta_time - x)/(delta_cool_err*delta_cool_err) * arg)
    return -np.sum((delta_time - x)/(delta_cool_err*delta_cool_err) * arg)
#    return a


    



# The prior functions
def ln_prior(p):
    bt1, th1, th2, th3 = p

    b1 = bt1 / np.cos(th1)

    # Get slopes from theta's
    m1 = np.tan(th1)
    m2 = np.tan(th2)
    m3 = np.tan(th3)
    
    # Determine b2, b3 as a function of other parameters
    b2 = (m1-m2) * 2.0 + b1
    b3 = (m2-m3) * 4.0 + b2
    
    lp = 0.
    
    # Limit the x-intercept: b = [0,0.5]
    if b1 < 0.:
#    if b1 < 0. or b1 > 0.5:
        return -np.inf
    
    # The slopes must all be positive
    # This corresponds to theta = [0,pi/2]
    if th1 < 0. or th1 > np.pi/2.:
        return -np.inf
    if th2 < 0. or th2 > np.pi/2.:
        return -np.inf
    if th3 < 0. or th3 > np.pi/2.:
        return -np.inf
    
    
    # Require that M_zams=7 becomes a WD (M_wd < 1.4)
    if 7.0*m3+b3 > 1.4 or 7.0*m3+b3 < 1.0:
        return -np.inf
    
    
    return lp

```

```python
# Calculate likelihood function
def ln_likelihood_simple(p_full, MESA_tcool, DWD, N_grid=200):
    
    p = p_full[0], p_full[1], p_full[2], p_full[3]
    t_sys = p_full[4:]
    
    # Limit the WD mass to be confined within this range
    wd_lower = 0.5    # Lowest allowed WD mass
    wd_upper = 1.4    # Highest allowed WD mass
    # First determine upper and lower
    zams_limits = np.array([min(MESA_tcool["name"]), max(MESA_tcool["name"])])
    wd_limits = get_m_wd(p, zams_limits)
    t_limits = func_lifetime(zams_limits)
    lower = wd_limits[0]
    upper = wd_limits[1]

    # Create probability array
    ln_P_sys = np.array([])
    
        
    # Calculate each system individually
#    for sys in DWD:
    for i in np.arange(len(DWD["m1_wd"])):
    
        sys = DWD[i]
        age_sys = t_sys[i]
        
        # wd mass limits
        # Go out to 5 sigma on either side
        wd1_min = max(lower, sys["m1_wd"] - 5.0*sys["m1_wd_err"])
        wd1_max = min(upper, sys["m1_wd"] + 5.0*sys["m1_wd_err"])
        wd2_min = max(lower, sys["m2_wd"] - 5.0*sys["m2_wd_err"])
        wd2_max = min(upper, sys["m2_wd"] + 5.0*sys["m2_wd_err"])

        # zams mass limits
        M_zams1_min, M_zams1_max = get_mzams(p, [wd1_min, wd1_max])    
        M_zams2_min, M_zams2_max = get_mzams(p, [wd2_min, wd2_max])    
    
        # lifetime limits
        t1_min = func_lifetime(M_zams1_min)
        t1_max = func_lifetime(M_zams1_max)
        t2_min = func_lifetime(M_zams2_min)
        t2_max = func_lifetime(M_zams2_max)

        # delta t limits
        delta_t_min = t1_max - t2_min
        delta_t_max = t1_min - t2_max

        wd1_grid = np.linspace(wd1_min, wd1_max, N_grid)
        wd2_grid = np.linspace(wd2_min, wd2_max, N_grid)    
    
        M1, M2 = np.meshgrid(wd1_grid, wd2_grid)
    
        # Grid width
        width_1 = (wd1_max - wd1_min) / N_grid
        width_2 = (wd2_max - wd2_min) / N_grid
        ln_w1 = np.log(width_1)
        ln_w2 = np.log(width_2)
        
        # M1
        ln_c1 = np.log(1.0/(sys["m1_wd_err"] * np.sqrt(2. * np.pi)))
        arg1 = -(sys["m1_wd"] - M1)*(sys["m1_wd"] - M1) / (2.0 * sys["m1_wd_err"] * sys["m1_wd_err"])

        # M2
        ln_c2 = np.log(1.0/(sys["m2_wd_err"] * np.sqrt(2. * np.pi)))
        arg2 = -(sys["m2_wd"] - M2)*(sys["m2_wd"] - M2) / (2.0 * sys["m2_wd_err"] * sys["m2_wd_err"])

        # delta t
        delta_cool = sys["age2"] - sys["age1"]
        delta_cool_err = np.sqrt(sys["age1_err"]*sys["age1_err"] + sys["age2_err"]*sys["age2_err"])

        M_zams1 = get_mzams(p, M1)
        M_zams2 = get_mzams(p, M2)
        t_1 = func_lifetime(M_zams1)
        t_2 = func_lifetime(M_zams2)
        delta_life = t_1 - t_2
        
        
#        ln_c3 = np.log(1./(delta_cool_err * np.sqrt(2. * np.pi)))
#        arg3 = -(delta_cool - delta_life)*(delta_cool - delta_life)/(2.0 * delta_cool_err*delta_cool_err)

        ln_c3 = np.log(1.0/(sys["age1_err"] * np.sqrt(2. * np.pi)))
        arg3 = -(sys["age1"] - (age_sys - t_1))*(sys["age1"] - (age_sys - t_1)) / (2.0 * sys["age1_err"] * sys["age1_err"])

        ln_c4 = np.log(1.0/(sys["age2_err"] * np.sqrt(2. * np.pi)))
        arg4 = -(sys["age2"] - (age_sys - t_2))*(sys["age2"] - (age_sys - t_2)) / (2.0 * sys["age2_err"] * sys["age2_err"])

        
        # Make priors flat in pre-wd lifetime, not mass
#        dt_dmzams1   = (func_lifetime(M_zams1+1.0e4) - func_lifetime(M_zams1)) / 1.0e4
#        dmzams_dmwd1 = (get_mzams(p, M1+1.0e-4) - get_mzams(p, M1)) / 1.0e-4
#        dt_dmzams2   = (func_lifetime(M_zams2+1.0e4) - func_lifetime(M_zams2)) / 1.0e4
#        dmzams_dmwd2 = (get_mzams(p, M2+1.0e-4) - get_mzams(p, M2)) / 1.0e-4
#  prior_m1 = dt_dmzams1 * dmzams_dmwd1 / (t_1+sys["age1"]) / np.log(t_limits[1]/t_limits[0])
#  prior_m2 = dt_dmzams2 * dmzams_dmwd2 / (t_2+sys["age2"]) / np.log(t_limits[1]/t_limits[0])
#        prior_m1 = dt_dmzams1 * dmzams_dmwd1 / (t1_max - t1_min)
#        prior_m2 = dt_dmzams2 * dmzams_dmwd2 / (t2_max - t2_min)

#        ln_p1 = np.log(prior_m1)
#        ln_p2 = np.log(prior_m2)

        
        
        # Calculate probability using logsumexp
#        ln_P_sys = np.append(ln_P_sys, logsumexp(ln_w1 + ln_w2 + ln_c1 + arg1 + ln_c2 + arg2 + ln_c3 + arg3)) 
#        ln_P_sys = np.append(ln_P_sys, logsumexp(ln_p1 + ln_p2 + ln_w1 + ln_w2 + ln_c1 + arg1 + ln_c2 + arg2 + ln_c3 + arg3))
        ln_P_sys = np.append(ln_P_sys, logsumexp(ln_w1 + ln_w2 + ln_c1 + arg1 + ln_c2 + arg2 + ln_c3 + arg3 + ln_c4 + arg4))
               
    return ln_P_sys 

# The posterior function calls the likelihood and prior functions
def ln_posterior_simple(p_full, MESA_tcool, DWD, N_grid=100):
    p = p_full[0], p_full[1], p_full[2], p_full[3]
    
    # Check prior on model first
    lp = ln_prior(p)
    
    if np.any(np.isnan(lp)):
        return -np.inf

    # Then check likelihood of parameters
    ll = ln_likelihood_simple(p_full, MESA_tcool, DWD, N_grid)
    
    if np.any(np.isnan(ll) | np.isinf(ll)):
        return -np.inf

    
    if np.all(ll == 0.):
        return -np.inf
    
    return lp + ll.sum()

```

### Wrapper for emcee

```python
# This function runs emcee
def run_inference(MESA_tcool, DWD, p_i, nwalkers=32, nburn=200, nsteps=1000):
    # number of parameters
    ndim = 4 + len(DWD["m1_wd"])

    # initial positions for walkers
    p0 = np.zeros((nwalkers,ndim))
    p0[:,0] = np.random.normal(p_i[0], 0.01, size=nwalkers) # b1
    p0[:,1] = np.random.normal(p_i[1], 0.01, size=nwalkers) # theta_1
    p0[:,2] = np.random.normal(p_i[2], 0.03, size=nwalkers) # theta_2
    p0[:,3] = np.random.normal(p_i[3], 0.03, size=nwalkers) # theta_3
    for i in np.arange(len(DWD["m1_wd"])):
        p0[:,4+i] = np.random.normal(2.0e9, 1.0e8, size=nwalkers)
    
    args = (MESA_tcool, DWD)
#    sampler = emcee.EnsembleSampler(nwalkers=nwalkers, dim=ndim, lnpostfn=ln_posterior, args=args)
    sampler = emcee.EnsembleSampler(nwalkers=nwalkers, dim=ndim, lnpostfn=ln_posterior_simple, args=args)


    # burn-in
    sampler.run_mcmc(p0, N=nburn)
    pos,prob,state = sampler.run_mcmc(p0, N=nburn)

    # throw out burn-in samples
    sampler.reset() 
    pos,prob,state = sampler.run_mcmc(pos, N=nsteps)

    
    return sampler
```
