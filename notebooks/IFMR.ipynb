### Creating the Model

```python
import os
import emcee
import triangle
from scipy.stats import truncnorm, norm
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


# Import MESA lifetimes
datafile = os.path.abspath("../data/MESA_lifetime.dat")
names = ["name","lifetime"]
MESA_tcool = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)

# Move to log10 of stellar ages
MESA_tcool["lifetime"] = np.log10(MESA_tcool["lifetime"])


# Import Wide DWD data
datafile = os.path.abspath("../data/WDWD_clean.dat")
names = ["name","m1_wd","m1_wd_err","age1","age1_err","m2_wd","m2_wd_err","age2","age2_err"]
DWD_andrews = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)

datafile = os.path.abspath("../data/Baxter_WDWD_clean.dat")
names = ["name","m1_wd","m1_wd_err","age1","age1_err","m2_wd","m2_wd_err","age2","age2_err"]
DWD_baxter = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)

# Combine our systems with those from Baxter
DWD = np.concatenate((DWD_andrews, DWD_baxter), axis=0)
    
```

```python

# Function to interpolate MESA lifetimes
def lifetime(MESA_tcool, M_zams):
    
    # Interpolation is cubic, in logspace over lifetime
    lifetime = interp1d(MESA_tcool["name"],MESA_tcool["lifetime"], kind='cubic', bounds_error=False)
    
    # Return linear version    
    return pow(10., lifetime(M_zams))



# Function to return M_zams from M_wd using model
def get_mzams(p, m_wds):
    b1, th1, th2, th3 = p
    
    # Get slopes from theta's
    m1 = np.tan(th1)
    m2 = np.tan(th2)
    m3 = np.tan(th3)
    
    M_zams = np.array([])

    # Determine b2, b3 as a function of other parameters
    b2 = (m1-m2) * 2.0 + b1
    b3 = (m2-m3) * 4.0 + b2
    

    # Define piecewise linear model
    for m_wd in m_wds:
        if m_wd > b1 and m_wd < 2.0*m1+b1:
            M_zams = np.append(M_zams, 1./m1 * (m_wd - b1))
    
        elif m_wd > 2.0*m1+b1 and m_wd < 4.0*m2+b2:
            M_zams = np.append(M_zams, 1./m2 * (m_wd - b2))
    
        elif m_wd > 4.0*m2+b2:
            M_zams = np.append(M_zams, 1./m3 * (m_wd - b3))
            
        else: 
            M_zams = np.append(M_zams, 0.)


    return M_zams


# Calculate likelihood function
def ln_likelihood(p, MESA_tcool, DWD):
    
    for sys in DWD:
    
    
    
        # We have to integrate over all possible M1_WD, M2_WD values
        # we'll do this numerically
        lower = 0.5    # Lowest allowed WD mass
        upper = 1.4    # Highest allowed WD mass
        m2s = np.linspace(lower, upper, 512)
        dm2 = m2s[1] - m2s[0]
        dist_m1 = truncnorm((lower - sys["m1_wd"]) / sys["m1_wd_err"], (upper - sys["m1_wd"]) / sys["m1_wd_err"], loc=sys["m1_wd"], scale=sys["m1_wd_err"])
        dist_m2 = truncnorm((lower - sys["m2_wd"]) / sys["m2_wd_err"], (upper - sys["m2_wd"]) / sys["m2_wd_err"], loc=sys["m2_wd"], scale=sys["m2_wd_err"])

        p_temp = norm(dist_m1)
        
        
        
    # Calculate zams masses
    m1_zams = get_mzams(p, DWD["m1_wd"])
    m2_zams = get_mzams(p, DWD["m2_wd"])
            
    # Calculate MS lifetimes
    m1_time = lifetime(MESA_tcool, m1_zams)
    m2_time = lifetime(MESA_tcool, m2_zams)
    
    # Now compare difference in lifetimes to difference in cooling ages
    delta_time = m1_time - m2_time
    delta_cool = DWD["age2"] - DWD["age1"]
    delta_cool_err = np.sqrt(DWD["age1_err"]*DWD["age1_err"] + DWD["age2_err"]*DWD["age2_err"])
    
    coeff = 1./(delta_cool_err * np.sqrt(2. * np.pi))
    argument = - (delta_time - delta_cool)*(delta_time - delta_cool)/(2. * delta_cool_err*delta_cool_err)
        
    return np.log(coeff) + argument   



# The prior functions
def ln_prior(p):
    b1, th1, th2, th3 = p
    
    lp = 0.
    
    # Limit the x-intercept: b = [0,0.5]
    if b1 < 0. or b1 > 0.5:
        return -np.inf
    
    # The slopes must all be positive
    # This corresponds to theta = [0,pi/2]
    
    if th1 < 0. or th1 > np.pi/2.:
        return -np.inf
    if th2 < 0. or th2 > np.pi/2.:
        return -np.inf
    if th3 < 0. or th3 > np.pi/2.:
        return -np.inf
    
    return lp


# The posterior function calls the likelihood and prior functions
def ln_posterior(p, MESA_tcool, DWD):
    
    ll = ln_likelihood(p, MESA_tcool, DWD)
    lp = ln_prior(p)
    
    if np.any(np.isnan(ll) | np.isinf(ll)):
        return -np.inf

    if np.any(np.isnan(lp)):
        return -np.inf
    
    if np.all(ll == 0.):
        return -np.inf
    
    return lp + ll.sum()

```

```python

# This function runs emcee
def run_inference(MESA_tcool, DWD, nwalkers=32, nburn=200, nsteps=1000):
    # number of parameters
    ndim = 4
    
    # initial positions for walkers
    p0 = np.zeros((nwalkers,ndim))
    p0[:,0] = np.random.normal(0.2, 0.05, size=nwalkers) # b1
    p0[:,1] = np.random.normal(0.15, 0.03, size=nwalkers) # theta_1
    p0[:,2] = np.random.normal(0.15, 0.03, size=nwalkers) # theta_2
    p0[:,3] = np.random.normal(0.15, 0.03, size=nwalkers) # theta_3
    
    args = (MESA_tcool, DWD)
    sampler = emcee.EnsembleSampler(nwalkers=nwalkers, dim=ndim, lnpostfn=ln_posterior, args=args)

    # burn-in
    sampler.run_mcmc(p0, N=nburn)
    pos,prob,state = sampler.run_mcmc(p0, N=nburn)

    # throw out burn-in samples
    sampler.reset() 
    pos,prob,state = sampler.run_mcmc(pos, N=nsteps)

    
    return sampler
```

```python
# initial positions for walkers
nwalkers = 32
ndim = 4

p0 = np.zeros((nwalkers,ndim))
p0[:,0] = np.random.normal(0.2, 0.05, size=nwalkers) # b1
p0[:,1] = np.random.normal(0.15, 0.03, size=nwalkers) # theta_1
p0[:,2] = np.random.normal(0.15, 0.03, size=nwalkers) # theta_2
p0[:,3] = np.random.normal(0.15, 0.03, size=nwalkers) # theta_3
```

```python
sampler = run_inference(MESA_tcool, DWD, nburn=200, nsteps=1000)
```

```python
labels = ["$b_1$", r"$\theta_1$", r"$\theta_2$", r"$\theta_3$"]

fig = triangle.corner(sampler.flatchain, labels=labels)
```
