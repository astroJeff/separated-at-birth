We are interested in finding the probability of the model ($\theta$) given the data ($D$):
    
\begin{equation}
P(\theta | D) = \frac{P(D | \theta) P(\theta)}{P(D)}
\end{equation}

Now we need to find the likelihood:

\begin{equation}
P(D | \theta) = P(\Delta t_{\rm cool}, M_{\rm 1,WD}, M_{\rm 2,WD} | \theta)
\end{equation}

Now we need to marginalize over the individual WD masses.

\begin{equation}
 P(D | \theta) = \int_0^{\infty} \int_0^{\infty} P(\Delta t_{\rm cool} | \theta, M_{\rm 1,WD}, M_{\rm 2,WD}) P(M_{\rm 1,WD} | {\hat M_{\rm 1,WD}}) P(M_{\rm 2,WD} | {\hat M_{\rm 2,WD}}) {\rm d}{\hat M_{\rm 1,WD}} {\rm d}{\hat M_{\rm 2,WD}}
\end{equation}

Given the measured WD mass ($M_{\rm WD}$), the probability of the true WD mass (${\hat M_{\rm WD}}$) is determined by a Gaussian with uncertainty $\sigma_{M_{\rm WD}}$:

\begin{equation}
P(M_{\rm WD} | {\hat M_{\rm WD}}) = \mathcal{N}(M_{\rm WD},\sigma_{M_{\rm WD}})
\end{equation}

Now we need to determine $P(\Delta t_{\rm cool} | \theta, M_{\rm 1,WD}, M_{\rm 2,WD})$. To do this, we marginalize over the difference in the main sequence lifetimes:

\begin{equation}
P(\Delta t_{\rm cool} | \theta, M_{\rm 1,WD}, M_{\rm 2,WD}) = \int_0^{\infty} P(\Delta t_{\rm cool} | \Delta t_{\rm MS}) P(\Delta t_{\rm MS} | \theta, M_{\rm 1,WD}, M_{\rm 2,WD}) {\rm d} \Delta t_{\rm MS}
\end{equation}

The probability $P(\Delta t_{\rm MS} | \theta, M_{\rm 1,WD}, M_{\rm 2,WD})$ becomes a delta function:

\begin{equation}
P(\Delta t_{\rm cool} | \theta, M_{\rm 1,WD}, M_{\rm 2,WD}) = \int_0^{\infty} \mathcal{N}(\Delta t_{\rm MS} | \Delta t_{\rm cool}, \sigma_{t_{\rm cool}}) \delta \left( \Delta t_{\rm MS} - f(\theta, M_{\rm 1,WD}, M_{\rm 2,WD}) \right) {\rm d} \Delta t_{\rm MS}
\end{equation}

where $f$ is some function of $\theta, M_{\rm 1,WD}, M_{\rm 2,WD}$. This equation is of the form:

\begin{equation}
\int F(x) \delta \left[ G(x) \right] {\rm d}x = \sum_i \frac{F(x_i)}{|G'(x_i)|}
\end{equation}

where the sum is over the $i$ roots of the function, $G(x)$, and the derivative $G'$ is with respect to the integrated variable, $x$. Our integral now reduces to:

\begin{equation}
P(\Delta t_{\rm cool} | \theta, M_{\rm 1,WD}, M_{\rm 2,WD}) = \frac{P(\Delta t_{\rm cool} | \Delta t_{\rm MS})}{\sum |g'(\theta, M_{\rm 1,WD}, M_{\rm 2,WD})|}
\end{equation}




These integrals reduce to:

\begin{equation}
P(\Delta t_{\rm cool} | \theta, M_{\rm 1,WD}, M_{\rm 2,WD}) = \frac{P(\Delta t_{\rm cool} | \theta, M_{\rm 1,ZAMS}, M_{\rm 2,ZAMS})} {|f'(M_{\rm 1,ZAMS})| |f'(M_{\rm 2,ZAMS})|}
\end{equation}

These delta functions are simply the inverse IFMR, therefore the function derivatives are the inverse of the slope of the IFMR model that we adopt. We model the IFMR as a piecewise linear function:

\begin{equation}
M_{\rm WD} = \left\{
  \begin{array}{ll}
    m_1 M_{\rm ZAMS} + b_1 & : M_{\rm ZAMS}/M_{\odot}<2.0 \\
    m_2 M_{\rm ZAMS} + b_2 & : 2.0<M_{\rm ZAMS}/M_{\odot}<4.0 \\
    m_3 M_{\rm ZAMS} + b_3 & : 4.0<M_{\rm ZAMS}/M_{\odot}
    \end{array}
    \right.
\end{equation}

To make the equation continuous, we solve for $b_2$ and $b_3$:
\begin{eqnarray}
b_2 &=& 2.0(m_1-m_2) + b_1 \\
b_3 &=& 2.0(m_1+m_2) - 4.0(m_3) + b_1
\end{eqnarray}

Our model therefore has 4 parameters, $b_1$, $m_1$, $m_2$, and $m_3$. From this model, we can determine the piecewise inverse IFMR:
\begin{equation}
M_{\rm ZAMS} = \left\{
  \begin{array}{ll}
    \frac{1}{m_1} M_{\rm WD} - b_1 & : b_1 < M_{\rm WD}/M_{\odot} < 2.0(m_1)+b_1 \\
    \frac{1}{m_2} M_{\rm WD} - b_2 & : 2.0(m_1)+b_1 < M_{\rm WD}/M_{\odot} < 4.0(m_2)+b_2 \\
    \frac{1}{m_3} M_{\rm WD} - b_3 & : 4.0(m_2)+b_2 < M_{\rm WD}/M_{\odot}
    \end{array}
    \right.
\end{equation}

Since our function derivatives are drawn from the same model IFMR, we can determine them both using the following piecewise function:

\begin{equation}
f'(M_{\rm ZAMS}) = \left\{
  \begin{array}{ll}
    \frac{1}{m_1} M_{\rm WD} - b_1 & : b_1 < M_{\rm WD}/M_{\odot} < 2.0(m_1)+b_1 \\
    \frac{1}{m_2} M_{\rm WD} - b_2 & : 2.0(m_1)+b_1 < M_{\rm WD}/M_{\odot} < 4.0(m_2)+b_2 \\
    \frac{1}{m_3} M_{\rm WD} - b_3 & : 4.0(m_2)+b_2 < M_{\rm WD}/M_{\odot}
    \end{array}
    \right.
\end{equation}



### Load the data

```python
import os
import emcee
import triangle
from scipy.stats import truncnorm, norm
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


# Import MESA lifetimes
datafile = os.path.abspath("../data/MESA_lifetime.dat")
names = ["name","lifetime"]
MESA_tcool = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)

# Move to log10 of stellar ages
MESA_tcool["lifetime"] = np.log10(MESA_tcool["lifetime"])


# Import Wide DWD data
datafile = os.path.abspath("../data/WDWD_clean.dat")
names = ["name","m1_wd","m1_wd_err","age1","age1_err","m2_wd","m2_wd_err","age2","age2_err"]
DWD_andrews = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)

datafile = os.path.abspath("../data/Baxter_WDWD_clean.dat")
names = ["name","m1_wd","m1_wd_err","age1","age1_err","m2_wd","m2_wd_err","age2","age2_err"]
DWD_baxter = np.genfromtxt(datafile, dtype=None, usecols=range(len(names)), names=names)

# Combine our systems with those from Baxter
DWD = np.concatenate((DWD_andrews, DWD_baxter), axis=0)
    
```

### Create function to interpolate between MESA lifetimes

```python
# Function to interpolate MESA lifetimes
def lifetime(MESA_tcool, M_zams):
    
    # Interpolation is cubic, in logspace over lifetime
    lifetime = interp1d(MESA_tcool["name"],MESA_tcool["lifetime"], kind='cubic', bounds_error=False, fill_value=7.0)
    
    # Return linear version    
    return pow(10., lifetime(M_zams))

```

### Functions for likelihood, prior, posterior, etc. for 3-piecewise linear model

```python

# Function to return M_zams from M_wd using model
def get_mzams(p, m_wds):
    b1, th1, th2, th3 = p
    
    # Get slopes from theta's
    m1 = np.tan(th1)
    m2 = np.tan(th2)
    m3 = np.tan(th3)
    
    # Determine b2, b3 as a function of other parameters
    b2 = (m1-m2) * 2.0 + b1
    b3 = (m2-m3) * 4.0 + b2
    
    # Define array
    M_zams = np.array([])

    # Define piecewise linear model
    for m_wd in m_wds:
        if m_wd > b1 and m_wd < 2.0*m1+b1:
            M_zams = np.append(M_zams, 1./m1 * (m_wd - b1))
    
        elif m_wd > 2.0*m1+b1 and m_wd < 4.0*m2+b2:
            M_zams = np.append(M_zams, 1./m2 * (m_wd - b2))
    
        elif m_wd > 4.0*m2+b2:
            M_zams = np.append(M_zams, 1./m3 * (m_wd - b3))
            
        else: 
            M_zams = np.append(M_zams, 0.)


    return M_zams


# Calculate likelihood function
def ln_likelihood(p, MESA_tcool, DWD):
    
    # Number of samples for our Monte Carlo integration
    N_samp = 100
    
    
    # Limit the WD mass to be confined within this range
    lower = 0.5    # Lowest allowed WD mass
    upper = 1.4    # Highest allowed WD mass
    
    # Create probability array
    P_sys = np.array([])

    # Calculate each system individually
    for sys in DWD:

        # Define array of cooling age differences
        delta_cool = sys["age2"] - sys["age1"]
        delta_cool_err = np.sqrt(sys["age1_err"]*sys["age1_err"] + sys["age2_err"]*sys["age2_err"])

        # Define truncated normal distributions for WD masses
        dist_wd1 = truncnorm.rvs((lower - sys["m1_wd"]) / sys["m1_wd_err"], (upper - sys["m1_wd"]) / sys["m1_wd_err"], loc=sys["m1_wd"], scale=sys["m1_wd_err"], size=N_samp)
        dist_wd2 = truncnorm.rvs((lower - sys["m2_wd"]) / sys["m2_wd_err"], (upper - sys["m2_wd"]) / sys["m2_wd_err"], loc=sys["m2_wd"], scale=sys["m2_wd_err"], size=N_samp)
                        
        # Calculate zams masses
        dist_zams1 = get_mzams(p, dist_wd1)
        dist_zams2 = get_mzams(p, dist_wd2)
        
        # Calculate MS lifetimes
        dist_time1 = lifetime(MESA_tcool, dist_zams1)
        dist_time2 = lifetime(MESA_tcool, dist_zams2)
    
        # Now compare difference in lifetimes to difference in cooling ages
        delta_time = dist_time1 - dist_time2
    
        # P( delta t_cool | M1_zams_t, M2_zams_t)
        coeff = 1./(delta_cool_err * np.sqrt(2. * np.pi))
        argument = np.exp(-(delta_time - delta_cool)*(delta_time - delta_cool)/(2. * delta_cool_err*delta_cool_err))
        
        # This is the value of the integrand
        P_sys_temp = coeff * argument

        P_sys = np.append(P_sys, np.sum(P_sys_temp))
                
                
    
    P_sys = P_sys/(N_samp + 0.0)
        
    ln_P_sys = np.log(P_sys)            
    ln_P_sys[np.isnan(ln_P_sys)] = -np.inf
    
    return ln_P_sys 



# The prior functions
def ln_prior(p):
    b1, th1, th2, th3 = p
    
    
    # Get slopes from theta's
    m1 = np.tan(th1)
    m2 = np.tan(th2)
    m3 = np.tan(th3)
    
    # Determine b2, b3 as a function of other parameters
    b2 = (m1-m2) * 2.0 + b1
    b3 = (m2-m3) * 4.0 + b2
    
    
    lp = 0.
    
    # Limit the x-intercept: b = [0,0.5]
    if b1 < 0. or b1 > 0.5:
        return -np.inf
    
    # The slopes must all be positive
    # This corresponds to theta = [0,pi/2]
    
    if th1 < 0. or th1 > np.pi/2.:
        return -np.inf
    if th2 < 0. or th2 > np.pi/2.:
        return -np.inf
    if th3 < 0. or th3 > np.pi/2.:
        return -np.inf
    
    
    # Require that M_zams=7 becomes a WD (M_wd < 1.4)
    if 7.0*m3+b3 > 1.4:
        return -np.inf
    
    
    return lp


# The posterior function calls the likelihood and prior functions
def ln_posterior(p, MESA_tcool, DWD):
    
    ll = ln_likelihood(p, MESA_tcool, DWD)
    lp = ln_prior(p)
    
    if np.any(np.isnan(ll) | np.isinf(ll)):
        return -np.inf

    if np.any(np.isnan(lp)):
        return -np.inf
    
    if np.all(ll == 0.):
        return -np.inf
    
    return lp + ll.sum()

```

### Wrapper for emcee

```python

# This function runs emcee
def run_inference(MESA_tcool, DWD, nwalkers=32, nburn=200, nsteps=1000):
    # number of parameters
    ndim = 4
    
    # initial positions for walkers
    p0 = np.zeros((nwalkers,ndim))
    p0[:,0] = np.random.normal(0.2, 0.05, size=nwalkers) # b1
    p0[:,1] = np.random.normal(0.15, 0.03, size=nwalkers) # theta_1
    p0[:,2] = np.random.normal(0.15, 0.03, size=nwalkers) # theta_2
    p0[:,3] = np.random.normal(0.15, 0.03, size=nwalkers) # theta_3
    
    args = (MESA_tcool, DWD)
    sampler = emcee.EnsembleSampler(nwalkers=nwalkers, dim=ndim, lnpostfn=ln_posterior, args=args)

    # burn-in
    sampler.run_mcmc(p0, N=nburn)
    pos,prob,state = sampler.run_mcmc(p0, N=nburn)

    # throw out burn-in samples
    sampler.reset() 
    pos,prob,state = sampler.run_mcmc(pos, N=nsteps)

    
    return sampler
```

### Run emcee

```python
sampler = run_inference(MESA_tcool, DWD, nburn=200, nsteps=500)
```
